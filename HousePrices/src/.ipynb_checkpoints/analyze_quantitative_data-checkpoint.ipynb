### Relevant imports and data load

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from scipy import stats

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer

%matplotlib inline
```

```python
# load data which is stored in the /data folder of the project
train_data = pd.read_csv('../data/train_filled_up.csv', sep=',', header=0)
```

```python
target_variable = train_data["SalePrice"]
```

## Analyze target variable

```python
target_variable.describe()
```

```python
sns.distplot(target_variable)
```

##### Outliers

```python
saleprice_scaled = StandardScaler().fit_transform(target_variable.values.reshape(-1,1))
low_range = saleprice_scaled[saleprice_scaled[:,0].argsort()][:10]
high_range = saleprice_scaled[saleprice_scaled[:,0].argsort()][-10:]
print('outer range (low) of the distribution')
print(low_range)
print('outer range (high) of the distribution')
print(high_range)
```

usually, threshold is set to -2.5 and +2.5 with small (80 or fewer) sample sizes
threshold is set -4.0 and +4.0 with large sample sizes
This means we have some outliers at the higher end in this case. Lower end seems to be ok. 

```python
sns.distplot(saleprice_scaled)
```

```python
sns.boxplot(target_variable)
```

##### Normality

```python
stats.kurtosis(target_variable)
```

```python
stats.skew(target_variable)
```

##### Linearity

```python
stats.probplot(target_variable, plot=plt)
```

##### Transformation

```python
log_target_variable = np.log(target_variable)
```

```python
sns.distplot(log_target_variable)
```

```python
stats.probplot(log_target_variable, plot=plt)
```

```python
sqrt_target_variable = np.sqrt(target_variable)
```

```python
sns.distplot(sqrt_target_variable)
```

```python
stats.probplot(sqrt_target_variable, plot=plt)
```

```python
print(log_target_variable.mean()/log_target_variable.std())
```

```python
print(sqrt_target_variable.mean()/sqrt_target_variable.std())
```

Given the probplot result and the ratio of mean and standard deviation, let's go with the log-transformation firts. 

```python
transformed_target_variable = log_target_variable
```
